```python
from mapper import mapper
import filters
import covers
import csv
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('more_points.csv') as points_file:
    reader = csv.reader(points_file)
    points_array = np.array([i for i in reader], dtype=float)

plt.scatter(points_array[:,0], points_array[:, 1])
```

```python
mapped = mapper(points=points_array, filter=filters.coordinate_filter(dims=[1]), cover=covers.uniform_cover(resolution=.1, gain=.25), epsilon=.1, neighbors=3, metric='euclidean')
mapped.graph_nerve()
```

```python
mapped = mapper(points=points_array, filter=filters.local_stat_filters(measure='mean', window_size=10, overlap=.1), cover=covers.uniform_cover(resolution=.1, gain=.25), epsilon=.1, neighbors=3, metric='euclidean')
mapped.graph_nerve()
```

```python
mapped = mapper(points=points_array, filter=filters.local_stat_filters(measure='std', window_size=5, overlap=.1), cover=covers.uniform_cover(resolution=.1, gain=.25), epsilon=.1, neighbors=3, metric='euclidean')
mapped.graph_nerve()
```

```python
mapped = mapper(points=points_array, filter=filters.density_filter(metric = 'euclidean', neighbors = 20, normalize=False), cover=covers.uniform_cover(resolution=.1, gain=.25), epsilon=.1, neighbors=3, metric='euclidean')
mapped.graph_nerve()
```

```python
mapped = mapper(points=points_array, filter=filters.density_filter(metric = 'geodesic', neighbors = 2, normalize=False), cover=covers.uniform_cover(resolution=.1, gain=.25), epsilon=.1, neighbors=3, metric='euclidean')
mapped.graph_nerve()
```
